```python
import tkinter as tk
from tkinter import messagebox
import numpy as np
from scipy import stats

class CalculatorApp:
    def __init__(self, root):
        self.root = root
        self.root.title("GUI Calculator")
        
        # Entry field for user input
        self.entry = tk.Entry(root, width=20, font=('Arial', 14))
        self.entry.grid(row=0, column=0, columnspan=4, padx=10, pady=10)
        
        # Buttons for numbers and operations
        buttons = [
            ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
            ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
            ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
            ('0', 4, 0), ('.', 4, 1), ('=', 4, 2), ('+', 4, 3),
        ]
        
        for btn_text, row, col in buttons:
            tk.Button(root, text=btn_text, width=5, height=2, font=('Arial', 12),
                      command=lambda text=btn_text: self.on_button_click(text)).grid(row=row, column=col, padx=5, pady=5)
        
        # Button for Clear
        tk.Button(root, text='Clear', width=5, height=2, font=('Arial', 12), command=self.clear_entry).grid(row=5, column=0, columnspan=2, padx=5, pady=5)
        
        # Button for Exit
        tk.Button(root, text='Exit', width=5, height=2, font=('Arial', 12), command=root.quit).grid(row=5, column=2, columnspan=2, padx=5, pady=5)
        
    def on_button_click(self, text):
        current_entry = self.entry.get()
        
        if text == '=':
            try:
                result = eval(current_entry)
                self.entry.delete(0, tk.END)
                self.entry.insert(tk.END, str(result))
            except Exception as e:
                messagebox.showerror("Error", f"Invalid input: {e}")
        
        elif text == 'Clear':
            self.clear_entry()
        
        else:
            self.entry.insert(tk.END, text)
    
    def clear_entry(self):
        self.entry.delete(0, tk.END)

def main():
    root = tk.Tk()
    app = CalculatorApp(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
